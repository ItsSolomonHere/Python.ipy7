```python
#1.1 Import Libraries
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set(style="whitegrid")
```

```python
#1.2 Load the Dataset

try:
    iris_df = pd.read_csv("Iris.csv")  # Use full path if file is not in the same directory
    print("✅ Dataset loaded successfully!")
except FileNotFoundError:
    print("❌ File not found. Check the path.")
except Exception as e:
    print(f"❌ An error occurred: {e}")
```

```python
#1.3 Inspect First Few Rows
iris_df.head()
```

```python
#1.4 Check Data Types and Missing Values

print("Data Types:\n", iris_df.dtypes)
print("\nMissing Values:\n", iris_df.isnull().sum())
```

```python
#1.5 Clean the Dataset (Drop Id, No missing values)
# Drop the Id column as it's not useful
iris_df.drop(columns=["Id"], inplace=True)
iris_df.head()
```

```python
iris_df.describe()
```

```python
2.2 Group by Species and Calculate Mean
#Task 2: Basic Data Analysis
iris_df.groupby("Species").mean(numeric_only=True)
```

```python
#2.3 Observations (run as markdown cell)
"""
- Iris-setosa has the smallest petal size.
- Iris-virginica generally has the largest dimensions across all features.
- Petal length and width are especially useful for distinguishing species.
"""
```

```python
# Simulate a line trend of SepalLengthCm over index
plt.figure(figsize=(10, 4))
plt.plot(iris_df.index, iris_df["SepalLengthCm"])
plt.title("Sepal Length Trend (Index-based)")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
sns.barplot(data=iris_df, x="Species", y="PetalLengthCm")
plt.title("Average Petal Length per Species")
plt.ylabel("Petal Length (cm)")
plt.xlabel("Species")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=iris_df, x="SepalWidthCm", bins=20, kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x="SepalLengthCm", y="PetalLengthCm", hue="Species")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```
